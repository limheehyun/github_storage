```python
# -1)코드
f = open('s.txt', 'r')
l = []
for line in f:
    l.append(line)
l.sort()
f.close()
with open('s1.txt', 'w') as f:
    f.writelines(l)
with open('s1.txt') as f:
    print(f.read())
```

```python
# -2)코드
f = open('s.txt', 'r')
l = [a for a in f]
f.close()
l.sort(key = lambda a: a[a.find(" ")+1]) #두번째 값 기준으로 정렬
with open('s2.txt', 'w') as f:
    f.writelines(l)
with open('s2.txt') as f:
    print(f.read())
```

```python
# -3)코드
f = open('s.txt', 'r')
text = ""
text2 = ""
cnt = 0
for a in f:
    text = text + a.replace("\n"," ")
for a in text:
    text2 = text2 + a
    if(a == " "):
        cnt=cnt+1
    if(cnt==3):
        text2 = text2 + "\n"
        cnt=0
with open('s3.txt', 'w') as f:
    f.writelines(text2)
with open('s3.txt') as f:
    print(f.read())
```

#### 2번 문제 풀이
* 변수 n에 입력값을 받는다.
* n=1 이면 반복문을 실행하여 ID 값을 입력받는다. 입력받은 ID와 기존 회원명단을 저장한 access.txt 파일을 열어 그 안에 저장된 ID 값을 비교해본다. text의 ":"문자 전까지가 ID에 해당하므로 슬라이싱으로 ID만 찾아서 비교한다.만약 중복된 ID면 안내메시지 출력하고 재입력 받을 수 있게 한다. 중복이 아니면 반복문을 종료시키고 나머지 pw,name,student 값을 입력받는다. 입력받은 ID, pw(암호화된), name, student 값을 text에 일정한 규칙을 따른 한 줄로 입력해준 뒤 다시 맨처음으로 돌아가게 해준다.
    - pw는 hashlib을 이용하여 암호화해주는데, hashlib.new('sha1')로 생성하고 update(pw.encode()) 로 인코딩된 pw 를 넣어준다. 그리고 hexdigest()로 암호화하여준다.

* n=3 이면 종료시킨다.

```python
#3번 문제
def frange(start=0, stop=0, step=0.1):
    if(stop==0):
        start, stop = stop, start #입력값이 stop 뿐이라면 두 값 바꿔줌
    l=[]
    a = start
    while(a < stop):
        l.append(float("{:.1f}".format(a))) #소수점 1자리까지만 출력
        a = a + step
    return l

print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

#### 3번 문제 풀이
* frange(start, stop=0, step=0.1) 로 기본 인수를 가지게 한다. 사용자가 만일 frange(0.5) 와 같이 1개의 값만 입력했다면 입력된 값은 stop할 값이라고 판단한다. 이 경우 start=0.5, stop=0.0 이 들어가 있기 때문에 두 값을 바꿔서 start=0.0, stop = 0.5가 되게 만들어준다. 그리고 임의의 변수에 start 값을 입력해주고 stop 값보다 커지기 전 까지 step값 만큼 계속 더해준다. 이 변수의 값들을 빈 리스트에 저장해주고, 반환한다.

```python
#4번 문제
def sum(*num):
    a = 0
    for i in num:
        a = a + i
    return a

print(sum(0))
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
```

#### 4번 문제 풀이
* sum(*num)으로 함수를 생성한다. *num은 가변인수를 받기 때문에 튜플로 생성된다. 이 튜플에 있는 여러개의 값을 반복문을 통해 모두 더해서 반환해준다.

```python
#5번 문제
def myinitial(s):
    l = s.split() #띄어쓰기별로 구분
    l = list(map(lambda x: x[0].upper(), l)) #맨 앞글자만 대문자로 만들어 저장
    t = "".join(l) #리스트의 글자를 띄어쓰기 없이 문자열로 결합
    return t

print(myinitial("as soon as possible"))
```

#### 5번 문제 풀이
* 문자열을 받으면 split으로 띄어쓰기별로 쪼개서 리스트에 저장해준다.
* l = list(map(lambda x: x[0].upper(), l)) 로 작성하여 첫글자만 남기고 대문자로 변환해준다.
* 대문자로 바뀐 첫글자들만 빈칸없이 문자열로 만든 뒤 반환해주고 함수를 종료한다.

```python
#6번 문제
def myfact(n, total=1):
    if(n==1):
        return total
    else:
        total = total*n
        return myfact(n-1, total)

print(myfact(10))
```

#### 문제 6번 풀이
* n의 값을 받고 n=1이면 1을 반환해준다.
* n이 1이 아니면 total에 n을 곱해주고 total의 값과 n-1의 값을 myfact 함수에 넣어준다.
* n=1이 될 때까지 반복해준다.
* 1000! 까지는 잘 작동했지만 10000! 부터는 컴퓨터가 멈출까싶어 작동정지했습니다.

```python
#8번 문제 (모범답안3 이용함)
import requests
import re
import string  
import pickle

def call(b):  
    req = requests.get(b)
    source = req.text
    text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S)
    text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)
    text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)
    text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)
    text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/]",' ', text4, 0, re.I|re.S)
    s=text5.split()
    dic={}
    html_len=""
    tag=0
    for i in source:
        if(i == '<'): #문자가 '<' 이면 tag =1 
            tag = 1
            html_len = html_len+i
        elif(i == '>'): #문자가 '>' 이면 tag = 0
            tag = 0
            html_len = html_len+i
        if(tag == 0): #순수 텍스트면 continue 
            continue
        elif(tag == 1): #그외(태그일때)는 html_len 에 추가
            html_len = html_len + i
    url_name = b[b.find("://")+3:] #url 이름
    
    for i in s:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    
    with open('stopwordlist_eng.txt', 'r') as f:
        l = f.read().split("\n") #영어불용어 불러와서 list에 저장
    for i in l:
        for j in list(dic):
            if(j.lower()==i):
                del dic[j] #영어불용어를 key로 가진 사전 삭제           
    with open('stopwordlist_kor.txt', 'r') as f:
        l = f.read().split('\n')
    for i in l:
        for j in list(dic):
            if(i==j.lower()[:len(i)] or i==j.lower()[0-len(i):]):
                dic[j.strip(i)]=dic[j] #한글 불용어 제거한 key,value로 수정
                del dic[j] #한글 불용어 제거하기 전 key,value 삭제
    if("" in dic):
        del dic[""]
    
    with open(url_name+".html.txt", 'w') as f:
        f.write(html_len) #html 저장
    with open(url_name+".word_frequency.pickle.txt", 'wb') as f:
        pickle.dump(dic, f) #사전객체 저장
        
    with open(url_name+".word_frequency.pickle.txt", 'rb') as f:
        t = pickle.load(f) #사전 가져오기
    dic_list = sorted(t.items(), key=lambda x: x[1], reverse=True) #사전 키 값 순서로 정렬
    print(dic_list[:3]) #그 중 제일 큰 3개만 출력

print('http://cse.koreatech.ac.kr')    
call('http://cse.koreatech.ac.kr')
print()

print('https://www.koreatech.ac.kr')
call('https://www.koreatech.ac.kr')
print()

print('http://www.naver.com')
call('http://www.naver.com')
print()

print('http://www.daum.net')
call('http://www.daum.net')
print()

print('http://www.nytimes.com')
call('http://www.nytimes.com')
```

```python
# 검색엔진
import pickle
search = input("검색어를 입력하세요: ")
old_search_list = search.split()

with open('stopwordlist_eng.txt', 'r') as f:
    l = f.read().split('\n')
for i in l:
    for j in range(len(old_search_list)):
        if(i == old_search_list[j].lower()):
            del old_search_list[j] #영어 불용어 제거
search_list=old_search_list
with open('stopwordlist_kor.txt', 'r') as f:
    l = f.read().split('\n')
for i in l:
    for j in range(len(old_search_list)):
        if(i==old_search_list[j].lower()[:len(i)] or i==old_search_list[j].lower()[0-len(i):]):
            search_list[j] = old_search_list[j].replace(i,"") #한글 불용어 제거

check_list = ['cse.koreatech.ac.kr.word_frequency.pickle.txt',
             'www.koreatech.ac.kr.word_frequency.pickle.txt',
             'www.naver.com.word_frequency.pickle.txt',
             'www.daum.net.word_frequency.pickle.txt',
             'www.nytimes.com.word_frequency.pickle.txt']
search_val=0;
wei=10; #가중치
weight_list={}
print('keyword = ', search_list)
for i in check_list:
    with open(i, 'rb') as f:
        dic = pickle.load(f)
    for j in search_list:
        for k in dic.keys():
            if(j.lower() == k.lower()):
                search_val = search_val + (wei*dic[k])
        wei = wei/10.0 # 다음 단어는 앞 단어보다 가중치가 1/10
    weight_list[i[:i.find('.word_frequency.pickle')]] = search_val #{url:검색가중치값} 으로 저장
    search_val=0
    wei=10
final_list=sorted(weight_list.items(), key=lambda x: x[1], reverse=True) #가중치 큰 대로 정렬
num=1
for i in final_list:
    print(num,": ",i[0])
    num = num+1
```

#### 8번 문제 풀이
* 1) 먼저 불용어 제거를 위해 영어와 한글 불용어를 파일로 저장해준다. 먼저 영어 불용어는 띄어쓰기로 구분 가능하기때문에 dic의 key와 일치하면 불용어이므로 바로 삭제해준다. 한글 불용어는 어두와 어미에 있을때만 삭제해준다. 삭제하면 key의 이름이 바뀌어야 하므로 불용어 제거한 단어를 key로 가지는 사전값을 추가해주고, 기존의 불용어 제거 전  단어를 key로 가지는 사전값은 삭제해준다. 한글 불용어 제거를 하면 빈칸을 key로 가지는 사전이 있을 수 있으므로 빈칸을 찾아서 삭제해준다.
* 2) URL은 항상 '://' 다음에 오는게 url 이므로 분석했던 '://' 다음이 문자열에서 몇번째인지 확인하여 그 위치부터 슬라이싱으로 url를 찾는다. HTML 소스는 '<'가 나오면 '>'가 나오기 전까지 문자열 저장하는 방식으로 하면 HTML 소스를 가진 문자열이 만들어지고, 그 문자열을 파일에 저장해주면 된다.
* 3) import pickle을 이용하여 사전정보를 이진파일로 저장해준다. 불러올 때도 이진파일로 불러오면 된다.
* 4) 분석하고자 하는 사이트를 call() 함수에 넣어주면 자동으로 HTML소스 파일과 사전파일이 만들어진다.
* 5) 사전을 불러와서 사전의 값의 내림차순으로 정렬해준다.(람다 함수 사용) 정렬된 사전의 앞에서부터 3개만 출력해주면 가장 많은 단어 3개를 알 수 있다.
* 6) 검색어를 받아서 1)에서 한 것처럼 동일하게 불용어를 제거해준다. 불용어가 제거된 단어들을 각 사이트의 사전에서 검색하여 가중치를 부여하는데, 검색시 앞에쓴 단어일수록 가중치가 10배 높게 설정했다.(ex- 파이썬 쉽게 = 파이썬:가중치10, 쉽게:가중치1) 그 가중치를 모두 더하고, 가중치가 가장 높은 사이트 순서대로 정렬해준 뒤 출력해준다. 만일 가중치가 동일하다면 사이트의 순서는 바뀌지 않는다.
* 7) 이 검색엔진이 완성도를 높일려면 단순히 단어의 빈도와 가중치로 순위를 매기는 것이므로 검색에 포함되지 않았어도 검색한 단어와 연관이 있는 단어를 사진 사이트도 알려주는것이 검색시 원하는 바를 찾기 더 좋지 않을까 생각해본다.
