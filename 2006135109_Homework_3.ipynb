```python
#문제.5
#for~in 리터럴
def addall_a(list):
    a = []
    for i in list:
        a.append(i)
    return print(sum(a))

#리스트내포
def addall_b(list):
    a = [i for i in list]
    return print(sum(a))

#for~in 리터럴
def addallodd_a(list):
    a = []
    for i in list:
        if(i % 2 == 1):
            a.append(i)
    return print(sum(a))

#리스트내포
def addallodd_b(list):
    a = [i for i in list if(i % 2 ==1)]
    return print(sum(a))

addall_a([1, 2, 3, 4, 5, 6, 7, 8, 9])
addall_b([1, 2, 3, 4, 5, 6, 7, 8, 9])
addallodd_a([1, 2, 3, 4, 5, 6, 7, 8, 9])
addallodd_b([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

- addall은 받은 list의 모든 값을 새로운 리스트에 입력하고 새로운 리스트의 모든값을 합해서 return 해주도록 작성했다.

```python
#문제.7
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
a = list(d.keys())
b = list(d.values())

a.sort() #키 오름차순
l = [(i, d[i]) for i in a]
print(l)

a.reverse() #키 내림차순(오름차순 자료 역순으로)
l = [(i, d[i]) for i in a]
print(l)

b.sort() #값 오름차순
l = [(j, i) for i in b for j in a if(d[j]==i)]
print(l)

b.reverse() #값 내림차순
l = [(j, i) for i in b for j in a if(d[j]==i)]
print(l)
```

#### 문제.7 풀이
사전 d의 키와 값을 각각 리스트 a와 b에 저장한다.
- 1) a.sort() 로 a에 저장된 키 값을 오름차순정렬한다. 그리고 리스트 l에 튜플형식으로 (a의 값, a를 키로 가지는 d의 값) = (d.key, d.value) 순서대로 저장한다. 저장순서는 a의 정렬된 값 순서이므로 리스트l에는 키의 오름차순된 값이 저장된다.
- 2) 오름차순 정렬된 a를 reverse로 역순정렬한다. 오름차순의 역순은 내림차순이므로 a는 내림차순정렬이 되어있고 이걸 1과 같이 리스트 l에 a값 순서대로 저장하면 내림차순정렬된 값이 저장된다.
- 3) b.sort()로 b에 저장된 value를 오름차순정렬한다. 리스트 l에 이중for문으로 b와 d[a]를 비교하여 같은 값일때만 키와 값을 저장하도록 한다. (b와 값이 일치하는 사전d의 키, b) = (d.key, d.value) 이므로 value의 오름차순으로 자료가 저장된다.
- 4) b.reverse() 로 오름차순된 자료를 역순정렬하여 내림차순으로 만들고 3과 같이 리스트 l에 값을 비교하여 키와 값을 내림차순으로 저장한다.

```python
#ACM-ICPC 문제 1673번
while(1):
    try:
        n, k = input().split()
    except:
        break
    n = int(n)
    k = int(k)
    total = n
    stamp = n
    while(stamp >= k):
        n = stamp//k
        stamp = stamp % k + n
        total = total + n
    print(total)
```

2. 설명

    - 입력받은 두개의 값을 int로 변환해준다.
    - total에 전체 치킨개수를 받고 stamp에 현재 가진 도장개수를 센다.
    - 도장의 개수 stamp가 k개보다 많으면 stamp/k 의 몫 = 치킨수, 나머지 = 도장수 로 계산하여 total에 치킨개수를 추가해준다.
    - 위의 계산을 stamp 숫자가 k개보다 많을때까지 반복해주고, 반복이 끝나면 총 치킨 수를 출력해준다.

```python
#Incremental Project 문제
import requests
import string

def delete_tag(a): #태그 제거하는 함수
    tag = 0 #tag가 1인동안은 문자들을 태그로 인식, 0인동안은 순수텍스트로 인식
    lan = "" #태그 제외한 문자열만 삽입할 새 문자열
    a = a.replace('<style>', '<', a.count('<style>')) #문자치환
    a = a.replace('</style>', '>', a.count('</style>'))
    a = a.replace('<script>', '<', a.count('<script>'))
    a = a.replace('</script>', '>', a.count('</script>'))   
    for i in a:
        if(i == '<'): #문자가 '<' 이면 tag+1 & lan에 빈칸 삽입 
            tag = 1
            lan = lan + " "
        elif(i == '>'): #문자가 '>' 이면 tag-1 & lan에 빈칸 삽입
            tag = 0
            lan = lan + " "
        elif(tag == 0): #순수 텍스트면 lan에 문자열 추가 
            lan = lan + i
        else: #그외(태그일때)는 continue
            continue
    return lan.split() #리스트형태로 반환

def delete_punc(a): #특수문자 제거 및 공백값 삭제
    l = [i.strip(string.punctuation) for i in a]
    l2 = [i for i in l if(i != '')]
    return l2

def make_dict(a): #사전으로 만드는 함수
    b = list(set(a))
    c = {i:a.count(i) for i in b}
    return c

req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

t = delete_tag(source) #t에 태그제거 함수 호출
t = delete_punc(t) #특수문자 제거 및 공백값 삭제
t = make_dict(t) #사전타입으로 만듬

print(len(t)) #t의 길이 = 사전의 크기
print(t)
#print(source)
```
