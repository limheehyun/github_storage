#### 2번 문제 풀이
* 다형성이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 이름의 함수 호출시에 다른 반응을 하는것을 말한다. 다형성을 이용하면 코딩의 양이 줄어들고 코드의 가독성이 높아진다.
* 하나의 예로 위와 같이 작성했다. weapon 이라는 클래스에 attack 이라는 함수를 정의하고 자식클래스로 sword와 punch 를 만들었다. 그 중 sword 클래스에는 attack 함수를 정의해주고 punch 클래스는 비워두었다. 이때 attack 함수를 실행하면 sword 클래스에서는 'slice'를 출력해주고 punch 에서는 'attack'을 출력해줌으로써 같은 이름의 함수이지만 다른 결과를 출력해주는걸 볼 수 있다.

```python
#3번 문제
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
    
    def __str__(self):
        return "[Count(step: " + str(self.step) + ")] " + str(self.count)
    
    def incr(self):
        return self.__init__(self.count+self.step, self.step)
    
    def __call__(self):
        return self.incr()
        
    def __add__(self, other):
        return Counter(self.count+other, self.step)
    
    def __sub__(self, other):
        return Counter(self.count-other, self.step)
    
    def __gt__(self, other):
        return self.count > other
    
    def __ge__(self, other):
        return self.count >= other
    
    def __lt__(self, other):
        return self.count < other
    
    def __le__(self, other):
        return self.count <= other
    
    def __eq__(self, other):
        return self.count == other
    
    def __ne__(self, other):
        return self.count != other
    
if __name__ == "__main__":
    c = Counter(10)
    d = Counter(10,2)
    print(c)
    print(d)
    
    c.incr()
    d.incr()
    
    print(c)
    print(d)
    
    c()
    d()
    print(c)
    print(d)
    
    c = c+5
    d = d-5
    print(c)
    print(d)
    
    print(c>10)
    print(d>10)
    print(c<10)
    print(d<10)
    print(c==17)
    print(d!=9)
```

#### 3번 문제 풀이
* 요구사항1 : __ init__() 함수 정의해준다. (self, count, step=1) 로 설정하여 step 값이 입력되지 않으면 1로 설정되게 해준다.
* 요구사항2 : __ str__() 함수 정의하여 print시 출력할 문자열을 지정해준다.
* 요구사항3 : incr() 함수를 만들어 호출되면 count+step 을 실행시켜서 count에 저장해준다.
* 요구사항4 : __ call__() 함수 정의해준다. call 함수가 실행되면 incr() 함수 실행될수 있도록 해준다.
* 요구사항5 : __ add__() 와 __ sub__() 함수 정의해서 덧셈과 뺄셈 가능하게 해준다.
* 요구사항6 : __ gt__(), __ ge__(), __ lt__(), __ le__(), __ eq__(), __ ne__() 여섯개의 함수 정의해서 비교연산 가능하게 해준다.

#### 4번 문제 풀이
* Myset 클래스에서 __ init__() 은 생성자로 리스트를 입력받아서 self 에 리스트값을 넣어준다. 그 뒤 eliminate_duplicate() 메소드에 리스트값을 넣어 실행시켜주는데, eliminate_duplicate() 는 staticmethod기 때문에 self를 입력하지 않는다. eliminate_duplicate() 에서는 받은 리스트에서 중복된 값을 제외한 리스트를 만들어주는데, 이 때 리스트명이 'l'인 리스트에 덮어쓰기 때문에 init에서 self에 들어가는 리스트도 중복이 제거된 리스트로 변경된다.
* __ str__() 에서는 중복된 값을 제거한 리스트를 출력해주는데, 리스트를 문자열로 바꾼 뒤 result라는 문자열에 합쳐준다. 그리고 완성된 문자열을 return 시켜주면 Myset 클래스를 받은 변수를 print 할 때 __ str__() 에서 작성된 문자열이 출력된다.

```python
#5번 문제
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self, other): #합집합
        for e in other: 
            if e not in self: #기존의 리스트에 없는 값인지 확인
                self.append(e) # 없는 값이면 추가해줌
        return self #최종 합집합 반환
    
    def __and__(self, other): #교집합
        s = []
        for e in self: 
            if e in other: #self리스트의 값이 other 리스트에 있는지 확인
                s.append(e) # 있으면 빈 리스트에 추가해줌
        self[:] = s #교집합리스트를 self 리스트에 넣어줌
        return self #최종 교집합 반환
    
    def __sub__(self, other): #차집합
        s = []
        for e in self:
            if e not in other: #self리스트의 값이 other 리스트에 없는지 확인
                s.append(e) # 없으면 빈 리스트에 추가해줌
        self[:] = s #차집합리스트를 self 리스트에 넣어줌
        return self #최종 차집합 반환
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s|t
    print(u)
    
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s&t
    print(u)
    
    s = MySet([1,2,3])
    t = MySet([3,4,5])
    u = s-t
    print(u)
```

#### 6번 문제 풀이
* class 내부에 선언하지 않았을 때는 파이썬 내부에 가지고 있는 연산자로 계산하기 때문에 len과 bool을 선언하지 않아도 사용할 수 있다.

```python
#7번 문제
import requests
import re
import string

class SearchEngine(list):
    def __init__(self,*l):
        for i in l:
            self.append(i)
        self.dic_list=[]
        self.total_dic={}
        for e in self:
            req = requests.get(e)
            source = req.text
            text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S)
            text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)
            text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)
            text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)
            text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/]",' ', text4, 0, re.I|re.S)
            s=text5.lower().split()
            dic={}

            for i in s:
                if i in dic:
                    dic[i] += 1
                else:
                    dic[i] = 1

            #불용어 제거 part
            with open('stopwordlist_eng.txt', 'r') as f:
                l = f.read().split("\n") #영어불용어 불러와서 list에 저장
            for i in l:
                for j in list(dic):
                    if(j.lower()==i):
                        del dic[j] #영어불용어를 key로 가진 사전 삭제           
            with open('stopwordlist_kor.txt', 'r') as f:
                l = f.read().split('\n')
            for i in l:
                for j in list(dic):
                    if(i==j.lower()[:len(i)] or i==j.lower()[0-len(i):]):
                        if(j.strip(i) != ""):
                            dic[j.strip(i)]=dic[j] #한글 불용어 제거한 key,value로 수정
                        del dic[j] #한글 불용어 제거하기 전 key,value 삭제
                        
            for i in dic.keys():
                if i in self.total_dic:
                    self.total_dic[i] = self.total_dic[i]+dic[i]
                else:
                    self.total_dic[i] = dic[i]
            self.dic_list.append(dic)
        
    def addUrl(self,l):
        self.append(l)
        self.__init__()
        
    def removeUrl(self,l):
        if l in self:
            del self.dic_list[self.index(l)]
            del self[self.index(l)]
        else:
            pass
        
    def listUrls(self):
        for e in self:
            print(e)
            
    def getWordsFrequency(self):
        print(self.total_dic)
            
        
    def getMaxFrequencyWords(self):
        if(len(self)==0):
            print("None")
        else:
            max_value = max(self.total_dic.values())
            for i in self.total_dic.keys():
                if self.total_dic[i] == max_value:
                    print(i)
        
    def searchUrlByWord(self, word):
        check={}
        for i in range(len(self.dic_list)):
            if word.lower() in self.dic_list[i]:
                check[self[i]] = self.dic_list[i][word.lower()]
        if(len(check)==0):
            for i in self:
                print(i)
        else:
            max_value = max(check.values())
            for i in check.keys():
                if(check[i]==max_value):
                    print(i)
                    
                    
if __name__ == "__main__":
    w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
    w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
    w3 = SearchEngine()

    print(w1)
    print(w2)
    print(w3)

    w1.addUrl('https://github.com')
    w3.addUrl('http://stackoverflow.com')
    print(w1)
    print(w3)
    
    w1.removeUrl('http://www.cnn.com')
    w3.removeUrl('http://stackoverflow.com')
    print(w1)
    print(w3)
    
    w1.listUrls()
    w2.listUrls()
    w1.getWordsFrequency()
    w3.getWordsFrequency()
    
    w1.getMaxFrequencyWords()
    w3.getMaxFrequencyWords()
    
    w1.searchUrlByWord("news")
    w2.searchUrlByWord("news")
```

#### 7번 문제 풀이
* __ init__() 메소드에 url 리스트를 self에 받고 url별로 단어빈도를 사전으로 만들어준다. 만들어진 사전은 dic_list 라는 리스트에 값으로 저장해준다. dic_list는 사전을 값으로 가지는 리스트가 된다.
* addUrl() 메소드는 입력받은 url을 기존 url 리스트에 추가해준다. 추가해주면서 __ init__() 메소드를 한번 더 실행하여 단어빈도리스트도 업데이트 해준다.
* removeUrl() 메소드는 입력받은 url이 url 리스트에 있으면 url 리스트와 dic_list 에서 해당하는 값을 함께 삭제해준다.
* listUrls() 메소드는 for문을 이용하여 self에 저장된 모든 리스트를 출력시켜준다.
* getWordsFrequency() 메소드는 url들의 단어출현빈도의 합을 사전화 해서 출력해주는데 __ init__()에서 dic_list를 만들면서 total_dic에 사전들의 값을 모두 합쳐주는 작업을 같이 했다. 따라서 total_dic 을 출력해주면 된다.
* getMaxFrequencyWords() 메소드는 total_dic 에서 값이 가장 큰 키를 출력해준다.
* searchUrlByWord() 메소드는 입력받은 문자열과 dic_list에 저장했던 사전들의 키를 비교해본다.만약에 같은 글자라면 check라는 빈 사전에 {url:word 빈도수} 로 저장해준다. 이렇게 만들어진 check 사전의 크기가 0이면 모든 url에 해당 단어가 없으므로 모든 url을 출력해준다. 만약 check 사전의 크기가 0이 아니면 가지고 있는 값중에 가장 큰 값을 찾고, 그 값을 가지고 있는 키를 출력해준다. 여기서 키는 url로 되어있기 때문에 결과적으로 유사도가 높은 url을 출력해주는게 된다.

```python
#8,9번 문제
class SearchEngineWithOrderedWebWords(SearchEngine):        
    def getWordsFrequency(self, reverse=False):
        new_dict = self.total_dic
        if(reverse==False):
            self.sort_list = sorted(new_dict.items(), key=lambda x: x[1], reverse=True)
        else:
            self.sort_list = sorted(new_dict.items(), key=lambda x: x[1], reverse=False)
        print(self.sort_list)
        self[:]=sorted(new_dict.items(), key=lambda x: x[1], reverse=True)
        
    def __iter(self):
        self.c_num=0
        return self
    
    def __next__(self):
        if(len(self.sort_list) < self.c_num):
            raise StopIteration
        else:
            s_list = self.sort_list[self.c_num]
            self.c_num += 1
            return s_list

if __name__ == "__main__":
    w4 = SearchEngineWithOrderedWebWords('http://www.naver.com', 'https://www.nate.com', 'https://github.com')
    w4.getWordsFrequency()
    w4.getWordsFrequency(reverse=True)
```
